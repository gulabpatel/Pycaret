# 1. Loading Dataset

```python
from pycaret.datasets import get_data
data = get_data('kiva')
```

# 2. Initialize Setup

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
from pycaret.nlp import *
nlp1 = setup(data, target = 'en', session_id=123, log_experiment=True, log_plots = True, experiment_name='kiva1')
```
